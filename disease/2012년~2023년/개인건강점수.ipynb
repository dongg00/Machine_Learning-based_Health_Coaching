```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE # SMOTE는 학습 시 사용되므로 여기서는 직접적인 역할 없음
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV
import json

# --- 모델, 스케일러, 피처, 평균값 로드 ---
def load_ml_assets():
    """머신러닝 모델, 스케일러, 피처 리스트, 학습 데이터 평균값을 로드합니다."""
    assets = {}
    try:
        assets['model_dm'] = joblib.load('For_HE_DM_HbA1c_calibrated_real.pkl')
        assets['scaler_dm'] = joblib.load('HE_DM_HbA1c_scaler_real.pkl')
        assets['features_dm'] = joblib.load('HE_DM_HbA1c_input_features_list_real.pkl')
        assets['train_means_dm'] = joblib.load('train_means_dm_real.pkl')

        assets['model_hp'] = joblib.load('For_HE_HP_calibrated_real.pkl')
        assets['scaler_hp'] = joblib.load('HE_HP_scaler_real.pkl')
        assets['features_hp'] = joblib.load('HE_HP_input_features_list_real.pkl')
        assets['train_means_hp'] = joblib.load('train_means_hp_real.pkl')

        assets['model_obesity'] = joblib.load('For_HE_obe_calibrated_real.pkl')
        assets['scaler_obesity'] = joblib.load('HE_obe_scaler_real.pkl')
        assets['features_obesity'] = joblib.load('HE_obe_input_features_list_real.pkl')
        assets['train_means_obesity'] = joblib.load('train_means_obe_real.pkl')
        # print("모든 모델 파일 로드 성공.")
    except FileNotFoundError as e:
        print(f"오류: 모델 파일을 찾을 수 없습니다. {e}. 모든 .pkl 파일이 올바른 디렉토리에 있는지 확인하세요.")
        exit()
    except Exception as e:
        print(f"오류: 모델 파일 로드 중 예상치 못한 오류 발생: {e}")
        exit()
    
    try:
        assets['model_scores'] = joblib.load('model_accuracies.pkl')
        # print("모델 정확도 로드 성공.")
    except FileNotFoundError:
        print("경고: 'model_accuracies.pkl' 파일을 찾을 수 없습니다. 모델 정확도가 0으로 처리됩니다.")
        assets['model_scores'] = {"당뇨병": 0.0, "고혈압": 0.0, "비만": 0.0}
    except Exception as e:
        print(f"모델 정확도 로드 중 오류 발생: {e}")
        assets['model_scores'] = {"당뇨병": 0.0, "고혈압": 0.0, "비만": 0.0}
    return assets

# --- 각 클래스의 설명 (예측 보고서 생성에 사용) ---
DIABETES_EXPLANATIONS = {
    0: "혈당 수준이 정상입니다. 당뇨병 위험이 낮습니다.",
    1: "혈당이 정상보다 높지만 당뇨병 경계형으로 간주됩니다. 생활습관 개선이 필요합니다.",
    2: "당뇨병으로 예측되었습니다. 전문가와 상담이 필요합니다."
}

HYPERTENSION_EXPLANATIONS = {
    0: "혈압이 정상 범위에 있습니다.",
    1: "혈압이 정상보다 높지만 고혈압 전단계로 간주됩니다. 건강 관리가 필요합니다.",
    2: "고혈압 전단계로, 생활습관 개선과 치료가 필요할 수 있습니다.",
    3: "고혈압으로 예측되었습니다. 전문가의 진단과 치료가 필요합니다."
}

OBESITY_EXPLANATIONS = {
    0: "저체중입니다. 체중 증가가 필요할 수 있습니다.",
    1: "정상 체중입니다. 유지 관리가 필요합니다.",
    2: "과체중입니다. 체중 감량을 고려할 수 있습니다.",
    3: "비만입니다. 체중 감량과 건강 관리가 필요합니다.",
    4: "고도 비만입니다. 전문가의 진단과 관리가 필요합니다.",
    5: "초고도 비만입니다. 전문가의 도움을 받아야 합니다."
}

# --- 핵심 지표 맵 (예측 보고서 생성에 사용) ---
TOP_FEATURES_MAP = {
    "당뇨병": {"primary": "HE_glu"},
    "고혈압": {"primary": "HE_sbp1", "secondary": "HE_dbp1"},
    "비만": {"primary": "HE_BMI"}
}

FEATURE_LABEL_MAP = {
    "HE_sbp1": "수축기 혈압",
    "HE_dbp1": "이완기 혈압",
    "HE_glu": "공복혈당",
    "HE_BMI": "BMI",
    "HE_HbA1c": "당화혈색소"
}

# --- 사용자 원시 입력값을 숫자형/매핑된 값으로 변환하는 함수 ---
def map_user_raw_input(raw_user_data):
    """
    설문지 응답 코드 형태의 사용자 입력값을 모델이 이해할 수 있는 숫자형/매핑된 값으로 변환합니다.
    """
    mapped_data = raw_user_data.copy()

    # BE3_31: 1주일간 걷기 일수 (1: 안 함, 2: 1일, ..., 8: 매일, 88: 비해당, 99: 모름/무응답)
    be3_31_raw = mapped_data.get('BE3_31')
    if be3_31_raw in [1, 88]: mapped_data['BE3_31'] = 0
    elif be3_31_raw == 2: mapped_data['BE3_31'] = 1
    elif be3_31_raw == 3: mapped_data['BE3_31'] = 2
    elif be3_31_raw == 4: mapped_data['BE3_31'] = 3
    elif be3_31_raw == 5: mapped_data['BE3_31'] = 4
    elif be3_31_raw == 6: mapped_data['BE3_31'] = 5
    elif be3_31_raw == 7: mapped_data['BE3_31'] = 6
    elif be3_31_raw == 8: mapped_data['BE3_31'] = 7
    else: mapped_data['BE3_31'] = np.nan # None 또는 99

    # BE3_33: 1회 걷는 시간(분) (숫자, 88/99 처리 필요)
    be3_33_raw = mapped_data.get('BE3_33')
    if be3_33_raw in [88, 99]: mapped_data['BE3_33'] = 0
    # None이면 그대로 None 유지, 다른 숫자면 그대로 사용

    # L_BR_FQ: 최근 1년 동안 1주 동안 아침식사 빈도 (1: 5~7회, 2: 3~4회, 3: 1~2회, 4: 안 함, 9: 모름/무응답)
    l_br_fq_raw = mapped_data.get('L_BR_FQ')
    if l_br_fq_raw in [None, 9]: mapped_data['L_BR_FQ'] = np.nan
    elif l_br_fq_raw == 1: mapped_data['L_BR_FQ'] = 6.0
    elif l_br_fq_raw == 2: mapped_data['L_BR_FQ'] = 3.5
    elif l_br_fq_raw == 3: mapped_data['L_BR_FQ'] = 1.5
    elif l_br_fq_raw == 4: mapped_data['L_BR_FQ'] = 0.0
    else: mapped_data['L_BR_FQ'] = np.nan

    # BD1_11: 1년간 음주빈도 (1: 안 마심, 2: 월1회미만, ..., 6: 주4회이상, 8: 비해당, 9: 모름/무응답)
    bd1_11_raw = mapped_data.get('BD1_11')
    if bd1_11_raw in [None, 9]: mapped_data['BD1_11'] = np.nan
    elif bd1_11_raw in [1, 8]: mapped_data['BD1_11'] = 0
    elif bd1_11_raw == 2: mapped_data['BD1_11'] = 6
    elif bd1_11_raw == 3: mapped_data['BD1_11'] = 12
    elif bd1_11_raw == 4: mapped_data['BD1_11'] = 42
    elif bd1_11_raw == 5: mapped_data['BD1_11'] = 130
    elif bd1_11_raw == 6: mapped_data['BD1_11'] = 286
    else: mapped_data['BD1_11'] = np.nan

    # BD2_1: 한 번에 마시는 음주량 (1: 1-2잔, 2: 3-4잔, ..., 5: 10잔 이상, 8: 비해당, 9: 모름/무응답)
    bd2_1_raw = mapped_data.get('BD2_1')
    if bd2_1_raw in [None, 8, 9]: mapped_data['BD2_1'] = np.nan
    elif bd2_1_raw == 1: mapped_data['BD2_1'] = 1.5
    elif bd2_1_raw == 2: mapped_data['BD2_1'] = 3.5
    elif bd2_1_raw == 3: mapped_data['BD2_1'] = 5.5
    elif bd2_1_raw == 4: mapped_data['BD2_1'] = 8.0
    elif bd2_1_raw == 5: mapped_data['BD2_1'] = 10.0
    else: mapped_data['BD2_1'] = np.nan

    # BS3_2: 일반 담배 하루 평균 흡연량 (숫자, 888/999 처리 필요)
    bs3_2_raw = mapped_data.get('BS3_2')
    if bs3_2_raw == 888: mapped_data['BS3_2'] = 0
    elif bs3_2_raw == 999: mapped_data['BS3_2'] = np.nan
    # None이거나 다른 숫자면 그대로 사용

    return mapped_data

# --- 입력값 유효성 검사 및 이상치 처리 로직 ---
def validate_and_clip_input(user_data, disease_name):
    """
    사용자 입력값의 유효성을 검사하고 비정상적인 범위를 클리핑합니다.
    경고 메시지를 반환합니다.
    """
    validated_data = user_data.copy()
    warning_messages = []

    if disease_name == "당뇨병":
        glu = validated_data.get('HE_glu')
        hba1c = validated_data.get('HE_HbA1c')
        if glu is not None and (glu < 10 or glu > 600):
            warning_messages.append(f"공복혈당({glu})은(는) 비정상적인 범위입니다. 실제 값인지 확인해 주세요.")
            validated_data['HE_glu'] = np.clip(glu, 10, 600)
        if hba1c is not None and (hba1c < 3 or hba1c > 20):
            warning_messages.append(f"당화혈색소({hba1c})은(는) 비정상적인 범위입니다. 실제 값인지 확인해 주세요.")
            validated_data['HE_HbA1c'] = np.clip(hba1c, 3, 20)

    elif disease_name == "고혈압":
        sbp = validated_data.get('HE_sbp1')
        dbp = validated_data.get('HE_dbp1')
        if sbp is not None and (sbp < 40 or sbp > 300):
            warning_messages.append(f"수축기 혈압({sbp}mmHg)은(는) 비정상적으로 극단적인 값입니다. 실제 값인지 확인해 주세요.")
            validated_data['HE_sbp1'] = np.clip(sbp, 40, 300)
        if dbp is not None and (dbp < 20 or dbp > 180):
            warning_messages.append(f"이완기 혈압({dbp}mmHg)은(는) 비정상적으로 극단적인 값입니다. 실제 값인지 확인해 주세요.")
            validated_data['HE_dbp1'] = np.clip(dbp, 20, 180)

    elif disease_name == "비만":
        bmi = validated_data.get('HE_BMI')
        ht = validated_data.get('HE_ht')
        wt = validated_data.get('HE_wt')
        if bmi is not None and (bmi < 10 or bmi > 60):
            warning_messages.append(f"BMI({bmi})은(는) 비정상적인 범위입니다. 실제 값인지 확인해 주세요.")
            validated_data['HE_BMI'] = np.clip(bmi, 10, 60)
        
        # 키와 몸무게로 BMI 계산 후 입력된 BMI와 비교
        if ht is not None and wt is not None and ht > 0:
            calculated_bmi = round(wt / ((ht/100)**2), 1)
            input_bmi_val = validated_data.get('HE_BMI')
            if input_bmi_val is None or abs(calculated_bmi - input_bmi_val) > 2:
                if input_bmi_val is not None:
                    warning_messages.append(f"입력된 키({ht}cm)와 몸무게({wt}kg)로 계산된 BMI({calculated_bmi})가 입력된 BMI({input_bmi_val})와 차이가 큽니다. 계산된 BMI로 대체합니다.")
                validated_data['HE_BMI'] = calculated_bmi

    return validated_data, warning_messages

# --- HbA1c 추정 함수 ---
def estimate_hba1c_from_glucose(glu):
    """공복혈당(glu)을 기반으로 당화혈색소(HbA1c)를 추정합니다."""
    return round((glu + 46.7) / 28.7, 1)

# --- 사용자 데이터를 모델 입력 형식으로 변환 ---
def prepare_for_prediction(user_data, features, scaler, train_means):
    """
    사용자 데이터를 pandas DataFrame으로 변환하고, 누락된 값을 채운 후 스케일링합니다.
    """
    if not isinstance(user_data, dict):
        raise TypeError("user_data는 딕셔너리여야 합니다.")
    
    user_df = pd.DataFrame([user_data]).reindex(columns=features)
    user_df = user_df.fillna(train_means)
    user_scaled = scaler.transform(user_df)
    return user_scaled


#---------------------------------------------------------------**질병예측분류**-------------------------------------------------------------------------
# --- 예측 함수 ---
def predict_disease_probabilities(user_scaled, model, stages):
    """
    스케일링된 사용자 데이터를 이용해 질병 확률을 예측합니다.
    """
    probs = model.predict_proba(user_scaled)[0]
    stage_index = np.argmax(probs)
    predicted_class_name = stages.get(stage_index, "Unknown")
    class_probs_formatted = [f"{probs[i]:.2f}" for i in range(len(probs))]
    
    # 정상 클래스의 인덱스를 찾아 해당 확률을 반환
    normal_class_index = 0
    if "비만" in stages.values() and predicted_class_name == "정상 체중": # 비만 모델의 정상 체중은 인덱스 1
        normal_class_index = list(stages.keys())[list(stages.values()).index("정상 체중")]
    
    prob_normal_class = float(probs[normal_class_index]) if normal_class_index < len(probs) else 0.0

    return {
        "probabilities": class_probs_formatted,
        "predicted_class_index": stage_index, # 0-indexed
        "class_name": predicted_class_name,
        "prob_normal_class": prob_normal_class # 정상 클래스 확률
    }

# --- 질병 예측 보고서 생성 함수 ---
def generate_disease_report(user_name, disease_name, prediction_result, user_validated_data, model_scores_dict):
    """
    각 질병에 대한 상세 예측 보고서를 콘솔에 출력하고, 요약 데이터를 반환합니다.
    """
    normal_class_index_map = {
        "당뇨병": 0, "고혈압": 0, "비만": 1
    }
    class_labels_map = {
        "당뇨병": ["정상", "경계형 당뇨", "당뇨병"],
        "고혈압": ["정상 혈압", "주의 혈압 단계", "고혈압 전단계", "고혈압"],
        "비만": ["저체중", "정상 체중", "과체중", "비만", "고도 비만", "초고도 비만"]
    }

    predicted_class_name = prediction_result['class_name']
    prob_normal_class = prediction_result['prob_normal_class']
    health_score = round(prob_normal_class * 100, 1)
    
    predicted_prob_raw = float(prediction_result['probabilities'][prediction_result['predicted_class_index']])
    predicted_prob_str = f"{predicted_prob_raw * 100:.0f}%"

    # print(f"\n=== {user_name}님의 {disease_name} 예측 결과 ===")

    # if prediction_result.get('warnings'):
        # print(f"⚠️ {disease_name} 관련 입력값 경고:")
    #     for warning in prediction_result['warnings']:
    #         print(f"   - {warning}")
    #     print("   (이 경고는 입력된 값이 일반적인 범위를 벗어나 모델 예측에 영향을 줄 수 있음을 나타냅니다.)\n")

    # print(f"1. {user_name}님의 현재 {disease_name} 관련 상태는 **{predicted_class_name}** 입니다.")
    # class_probs_str = ', '.join([f'{label} 확률: {float(prob) * 100:.0f}%' for label, prob in zip(class_labels_map.get(disease_name, []), prediction_result['probabilities'])])
    # print(f"   ({class_probs_str})\n")
    # print(f"2. 건강 점수: {health_score:.1f}점 (이 점수는 해당 질병에 대한 '정상' 상태일 확률을 나타냅니다.)\n")

    # 주요 지표 정보
    primary_feature_info = {}
    if disease_name == "고혈압":
        sbp_val = user_validated_data.get(TOP_FEATURES_MAP["고혈압"]["primary"])
        dbp_val = user_validated_data.get(TOP_FEATURES_MAP["고혈압"]["secondary"])

        sbp_status = ""
        if sbp_val is not None:
            if sbp_val < 120: sbp_status = "정상"
            elif 120 <= sbp_val <= 129: sbp_status = "주의혈압"
            elif 130 <= sbp_val <= 139: sbp_status = "위험 (고혈압1기)"
            else: sbp_status = "고위험 (고혈압2기)"
        primary_feature_info["수축기 혈압"] = {"값": sbp_val, "상태": sbp_status, "정상_기준": "<120mmHg"}
        
        dbp_status = ""
        if dbp_val is not None:
            if dbp_val < 80: dbp_status = "정상"
            elif 80 <= dbp_val <= 89: dbp_status = "위험 (고혈압1기)"
            else: dbp_status = "고위험 (고혈압2기)"
        primary_feature_info["이완기 혈압"] = {"값": dbp_val, "상태": dbp_status, "정상_기준": "<80mmHg"}

        # print(f"3. {FEATURE_LABEL_MAP['HE_sbp1']} 수치: {sbp_val}mmHg → 상태: {sbp_status}")
        # print(f"   (정상 <120, 주의 120-129, 위험 130-139, 고위험 ≥140)")
        # print(f"   {FEATURE_LABEL_MAP['HE_dbp1']} 수치: {dbp_val}mmHg → 상태: {dbp_status}")
        # print(f"   (정상 <80, 위험 80-89, 고위험 ≥90)\n")
        # print(f"   {disease_name}과 가장 높게 연관된 수축기/이완기 혈압은(는) 높아지면 고혈압 위험이 증가합니다.\n")

    elif disease_name == "당뇨병":
        glu_val = user_validated_data.get(TOP_FEATURES_MAP["당뇨병"]["primary"])
        glu_status = ""
        if glu_val is not None:
            if glu_val <= 99: glu_status = "정상"
            elif 100 <= glu_val <= 125: glu_status = "주의"
            else: glu_status = "고위험"
        primary_feature_info[FEATURE_LABEL_MAP['HE_glu']] = {"값": glu_val, "상태": glu_status, "정상_기준": "70~99"}

        # print(f"3. {FEATURE_LABEL_MAP['HE_glu']} 수치: {glu_val} → 상태: {glu_status}")
        # print(f"   (정상 기준: 70~99, 주의 기준: 100~125)\n")
        # print(f"   {disease_name}과 가장 높게 연관된 {FEATURE_LABEL_MAP['HE_glu']}은(는) 높아지면 {disease_name} 위험이 증가합니다.\n")

    elif disease_name == "비만":
        bmi_val = user_validated_data.get(TOP_FEATURES_MAP["비만"]["primary"])
        bmi_status = ""
        if bmi_val is not None:
            if bmi_val < 18.5: bmi_status = "저체중"
            elif 18.5 <= bmi_val <= 22.9: bmi_status = "정상 체중"
            elif 23.0 <= bmi_val <= 24.9: bmi_status = "과체중"
            elif 25.0 <= bmi_val <= 29.9: bmi_status = "비만 1단계"
            elif 30.0 <= bmi_val <= 34.9: bmi_status = "비만 2단계"
            else: bmi_status = "초고도 비만"
        primary_feature_info[FEATURE_LABEL_MAP['HE_BMI']] = {"값": bmi_val, "상태": bmi_status, "정상_기준": "18.5~22.9"}
        
        # print(f"3. {FEATURE_LABEL_MAP['HE_BMI']} 수치: {bmi_val} → 상태: {bmi_status}")
        # print(f"   (정상 기준: 18.5~22.9, 과체중: 23.0~24.9, 비만: ≥25.0)\n")
        # print(f"   {disease_name}과 가장 높게 연관된 {FEATURE_LABEL_MAP['HE_BMI']}은(는) 높아지면 {disease_name} 위험이 증가합니다.\n")
    
    model_accuracy_str = f"{model_scores_dict.get(disease_name, 0.0):.1f}%"

    return {
        "질병": disease_name,
        "예측_상태": predicted_class_name,
        "예측_확률": predicted_prob_str,
        "정상_확률": f"{prob_normal_class * 100:.1f}%",
        "모델_정확도": model_accuracy_str,
        "주요_지표": primary_feature_info,
        "건강_점수": health_score # 건강 점수도 리턴
    }

    #-------------------------------------------------------------------**생활 습관별 점수화**---------------------------------------------------------------------------
    # --- 생활 습관 점수화 함수들 (100점 만점) ---
def calculate_smoking_score(bs3_2):
    """일반 담배 하루 평균 흡연량에 따른 점수 (38점 만점)"""
    # 999는 "모름/무응답"으로 간주
    if bs3_2 is None or pd.isna(bs3_2) or bs3_2 == 999:
        return 12.5 # 값이 없는 경우 중간 점수 (38 / 3)
    if bs3_2 == 0:  # 비흡연자 (888도 여기에 포함)
        return 38.0 
    elif bs3_2 <= 10: # 10개비 이하
        ftnd_score_level = 0
    elif bs3_2 <= 20: # 20개비 이하
        ftnd_score_level = 1
    elif bs3_2 <= 30: # 30개비 이하
        ftnd_score_level = 2
    else: # 30개비 초과
        ftnd_score_level = 3
    
    # 0점부터 38점까지 스케일링 (최저 0점, 최고 38점)
    # (3 - level) / 3 * max_score
    return round(38.0 * (3 - ftnd_score_level) / 3, 1) # 38점 만점으로 스케일 조정

def calculate_walking_score(be3_33, be3_31):
    """걷기 시간 및 일수에 따른 점수 (28점 만점)"""
    if be3_31 is None or be3_33 is None or pd.isna(be3_31) or pd.isna(be3_33):
        return 14.0 # 둘 중 하나라도 없으면 중간 점수 (28 / 2)
    
    # '안 함'이나 '비해당'이 매핑되어 0이 되는 경우 (map_user_raw_input에서 처리됨)
    # be3_31이 0이거나 be3_33이 0이면, 걷기 활동이 없는 것으로 간주
    if be3_31 == 0 or be3_33 == 0:
        return 0.0 
    
    met = (be3_33) * 3.3 * be3_31 # METs 계산

    # IPAQ 활동 수준 → 원점수 (1: 낮음, 2: 보통, 3: 높음)
    if met >= 3000:
        base_score_level = 3 # 매우 활발
    elif met >= 600:
        base_score_level = 2 # 충분히 활발
    else:
        base_score_level = 1 # 어느 정도 활동
    
    # 0점부터 28점까지 스케일링 (level이 1일 때 0점, level이 3일 때 28점)
    # (level - 1) / 2 * max_score
    return round(28.0 * (base_score_level - 1) / 2, 1)

def calculate_alcohol_score(bd1_11, bd2_1):
    """음주 빈도 및 양에 따른 점수 (20점 만점)"""
    # Null/NaN 값 처리 (중간 점수 반환)
    if bd1_11 is None or bd2_1 is None or pd.isna(bd1_11) or pd.isna(bd2_1):
        return 10.0 # 20점 만점에 절반 (중간 점수)

    # 음주 빈도 점수 (낮을수록 건강) - New logic
    if bd1_11 == 0: # 안 마심 (맵핑된 값 기준)
        freq_score = 0
    elif bd1_11 <= 12:  # 월1회 미만 또는 월1회정도 (맵핑된 값 기준)
        freq_score = 0 # Your base treats <=12 as the same as 0.0 for this score part
    elif bd1_11 <= 42:  # 월2~4회 (맵핑된 값 기준)
        freq_score = 1
    else:  # 주2회 이상 (맵핑된 값 기준)
        freq_score = 2

    # 1회 음주량 점수 (낮을수록 건강) - New logic
    # The base refers to '잔', so it's likely expecting the raw BD2_1 before mapping to 1.5, 3.5, etc.
    # However, the overall pipeline maps BD2_1 to 1.5, 3.5, etc.
    # I'll interpret "bd2_1 <= 2" as roughly equivalent to the original's 1.5 (1-2잔) and adjust
    # subsequent thresholds to match the mapped values.
    # Assuming bd2_1 here refers to the *mapped* values (1.5, 3.5, 5.5, 8.0, 10.0)
    # If the intent was to use original raw values (1,2,3,4,5), the mapping in map_user_raw_input needs to be skipped
    # for this function or the raw value needs to be passed.
    # Sticking to the current pipeline, bd2_1 will be the mapped value.
    if bd2_1 <= 1.5: # Corresponds to 1-2잔 (original mapped 1.5)
        amount_score = 0
    elif bd2_1 <= 3.5: # Corresponds to 3-4잔 (original mapped 3.5)
        amount_score = 1
    else: # Corresponds to 5-6잔 or more (original mapped >=5.5)
        amount_score = 2

    # 총 위험 레벨 (0~4) - New logic
    total_risk_level = freq_score + amount_score

    # 총점 기반 건강 점수 (1~3점) - New logic
    if total_risk_level <= 1:
        base_score_level = 3 # Low risk, high score
    elif total_risk_level <= 3:
        base_score_level = 2
    else: # total_risk_level == 4
        base_score_level = 1 # High risk, low score

    # 20점까지 스케일링 (level이 1일 때 0점, level이 3일 때 20점)
    return round(20.0 * (base_score_level - 1) / 2, 1)

def calculate_breakfast_score(l_br_fq):
    """아침식사 빈도에 따른 점수 (15점 만점)"""
    if l_br_fq is None or pd.isna(l_br_fq): 
        return 7.5 # 값이 없으면 중간 점수 (15 / 2)
    
    # 빈도 기반 점수화 (높을수록 건강) - New logic (using mapped values)
    # l_br_fq values: 0.0 (안 함), 1.5 (1-2회), 3.5 (3-4회), 6.0 (5-7회)
    if l_br_fq >= 6.0: # 주 5~7회 (맵핑된 값 기준)
        base_score_level = 3  # Daily
    elif l_br_fq >= 3.5: # 주 3~4회 (맵핑된 값 기준)
        base_score_level = 2  # Frequent
    else: # 0.0 (안함) or 1.5 (1-2회)
        base_score_level = 1  # Infrequent or Not at all
    
    # 0점부터 15점까지 스케일링 (level이 1일 때 0점, level이 3일 때 15점)
    # (base_score_level - 1) / 2 * max_score
    return round(15.0 * (base_score_level - 1) / 2, 1)


#**메인 실행 로직**
# --- 사용자 데이터 (기존과 동일) ---
users = {
    "홍길동": {
        "HE_glu": 100.0, "HE_HbA1c": 10.0, "HE_ht": 170, "HE_wt": 63, "HE_wc": 78,
        "HE_BMI": 21.8, "HE_sbp1": 135, "HE_dbp1": 85, "HE_Upro": 3, "HE_chol": 200,
        "HE_HDL_st2": 55, "HE_TG": 120, "HE_LDL_drct": 110, "HE_crea": 0.85,
        "HE_ast": 22, "HE_alt": 18, "age": 20,
        "BS3_2": 0,       # 일반 담배 하루 평균 흡연량 (숫자, 888/999 처리 필요)
        "BE3_33": 30,     # 1회 걷는 시간(분) (숫자, 88/99 처리 필요)
        "BE3_31": 8,      # 1주일간 걷기 일수 (설문 코드: 8은 매일)
        "BD1_11": 3,      # 1년간 음주빈도 (설문 코드: 3은 월1회정도)
        "BD2_1": 2,       # 한 번에 마시는 음주량 (설문 코드: 2는 3-4잔)
        "L_BR_FQ": 1      # 아침식사 빈도 (설문 코드: 1은 5~7회)
    },
    "김철수": {
        "HE_glu": 92.0, "HE_HbA1c": 5.5, "HE_ht": 178, "HE_wt": 75, "HE_wc": 85,
        "HE_BMI": 23.7, "HE_sbp1": 118, "HE_dbp1": 76, "HE_Upro": 1, "HE_chol": 185,
        "HE_HDL_st2": 58, "HE_TG": 90, "HE_LDL_drct": 95, "HE_crea": 0.9,
        "HE_ast": 20, "HE_alt": 19, "age": 28,
        "BS3_2": 10,
        "BE3_33": 60,
        "BE3_31": 6,
        "BD1_11": 4,
        "BD2_1": 3,
        "L_BR_FQ": 2
    }
}

# ----------------------------------------------------------------- 메인 실행 -----------------------------------------------------------------
if __name__ == "__main__":
    ml_assets = load_ml_assets()

    user_results = {}

    for user_name, user_data_original in users.items():
        # print(f"\n--- {user_name}님의 건강 및 생활습관 분석 시작 ---")

        # 1. 원시 사용자 데이터를 매핑
        user_data_mapped = map_user_raw_input(user_data_original)
        
        # 2. 당화혈색소(HbA1c) 추정 (공복혈당이 있고 HbA1c가 없을 경우)
        if ('HE_HbA1c' not in user_data_mapped or user_data_mapped['HE_HbA1c'] is None) \
           and ('HE_glu' in user_data_mapped and user_data_mapped['HE_glu'] is not None and user_data_mapped['HE_glu'] > 0):
            estimated_hba1c = estimate_hba1c_from_glucose(user_data_mapped['HE_glu'])
            user_data_mapped['HE_HbA1c'] = estimated_hba1c
            print(f"{user_name}: 당화혈색소(HbA1c) 수치가 없어 공복혈당을 기반으로 {estimated_hba1c}%로 추정했습니다.")
        elif ('HE_HbA1c' not in user_data_mapped or user_data_mapped['HE_HbA1c'] is None) and \
             ('HE_glu' not in user_data_mapped or user_data_mapped['HE_glu'] is None):
            print(f"{user_name}: HbA1c와 공복혈당 모두 없어 평균값으로 대체됩니다.") # fillna가 처리

        # 3. 질병 예측 및 건강 점수 계산
        current_health_scores_list = []
        disease_summary_for_user = []
        individual_health_scores = {} # Dictionary to store individual health scores

        # 당뇨병 예측
        user_data_dm_validated, warnings_dm = validate_and_clip_input(user_data_mapped, "당뇨병")
        user_scaled_dm = prepare_for_prediction(user_data_dm_validated, ml_assets['features_dm'], ml_assets['scaler_dm'], ml_assets['train_means_dm'])
        result_dm = predict_disease_probabilities(user_scaled_dm, ml_assets['model_dm'], {0: "정상", 1: "경계형 당뇨", 2: "당뇨병"})
        result_dm['warnings'] = warnings_dm
        report_dm = generate_disease_report(user_name, "당뇨병", result_dm, user_data_dm_validated, ml_assets['model_scores'])
        current_health_scores_list.append(report_dm['건강_점수'])
        disease_summary_for_user.append(report_dm)
        individual_health_scores["당뇨병_건강_점수"] = report_dm['건강_점수']


        # 고혈압 예측
        user_data_hp_validated, warnings_hp = validate_and_clip_input(user_data_mapped, "고혈압")
        user_scaled_hp = prepare_for_prediction(user_data_hp_validated, ml_assets['features_hp'], ml_assets['scaler_hp'], ml_assets['train_means_hp'])
        result_hp = predict_disease_probabilities(user_scaled_hp, ml_assets['model_hp'], {0: "정상 혈압", 1: "주의 혈압 단계", 2: "고혈압 전단계", 3: "고혈압"})
        result_hp['warnings'] = warnings_hp
        report_hp = generate_disease_report(user_name, "고혈압", result_hp, user_data_hp_validated, ml_assets['model_scores'])
        current_health_scores_list.append(report_hp['건강_점수'])
        disease_summary_for_user.append(report_hp)
        individual_health_scores["고혈압_건강_점수"] = report_hp['건강_점수']

        # 비만도 예측
        user_data_obesity_validated, warnings_obesity = validate_and_clip_input(user_data_mapped, "비만")
        user_scaled_obesity = prepare_for_prediction(user_data_obesity_validated, ml_assets['features_obesity'], ml_assets['scaler_obesity'], ml_assets['train_means_obesity'])
        result_obesity = predict_disease_probabilities(user_scaled_obesity, ml_assets['model_obesity'], {0: "저체중", 1: "정상 체중", 2: "과체중", 3: "비만", 4: "고도비만", 5: "초고도 비만"})
        result_obesity['warnings'] = warnings_obesity
        report_obesity = generate_disease_report(user_name, "비만", result_obesity, user_data_obesity_validated, ml_assets['model_scores'])
        current_health_scores_list.append(report_obesity['건강_점수'])
        disease_summary_for_user.append(report_obesity)
        individual_health_scores["비만_건강_점수"] = report_obesity['건강_점수']


        # 4. 생활 습관 점수 계산 (이 부분을 주석 처리하거나 제거)
        # smoking_score = calculate_smoking_score(user_data_mapped.get('BS3_2'))
        # walking_score = calculate_walking_score(user_data_mapped.get('BE3_33'), user_data_mapped.get('BE3_31'))
        # alcohol_score = calculate_alcohol_score(user_data_mapped.get('BD1_11'), user_data_mapped.get('BD2_1'))
        # breakfast_score = calculate_breakfast_score(user_data_mapped.get('L_BR_FQ'))
        
        # current_lifestyle_scores = []
        # current_lifestyle_scores.append(smoking_score)
        # current_lifestyle_scores.append(walking_score)
        # current_lifestyle_scores.append(alcohol_score)
        # current_lifestyle_scores.append(breakfast_score)

        # 5. 최종 점수 집계 및 저장 (생활습관 점수 관련 부분 수정)
        avg_health_score = round(sum(current_health_scores_list) / len(current_health_scores_list), 1)
        
        # 생활 습관 점수를 사용하지 않으므로 0으로 설정하거나 제거
        avg_lifestyle_score = 0.0 # 생활습관 점수를 사용하지 않음
        
        combined_total_score = avg_health_score # 건강점수만으로 총점 계산

        user_results[user_name] = {
            "건강_평균_점수": avg_health_score,
            "총_건강_점수": combined_total_score,
            "건강_세부_점수": individual_health_scores # Add the detailed health scores here
        }
        # print(f"\n--- {user_name}님의 분석 요약 ---")
        # print(f"건강 평균 점수 (정상 확률 기반): {avg_health_score}점")
        # # print(f"생활습관 평균 점수 (각 항목 가중치 반영 합산): 생활습관 점수 미고려") # 출력 메시지 변경
        # print(f"최종 개인 건강 점수 (정상 확률 기반): {combined_total_score}점\n")

    # --- 승자 결정 및 최종 결과 출력 ---
    sorted_users = sorted(user_results.items(), key=lambda x: x[1]["총_건강_점수"], reverse=True)

    print("\n--- 최종 결과 ---")
    # print("사용자별 최종 개인 건강 점수:")
    for user_name, result_data in sorted_users:
        print(f"- {user_name}: {result_data['총_건강_점수']}점")

    if len(sorted_users) > 1 and sorted_users[0][1]["총_건강_점수"] == sorted_users[1][1]["총_건강_점수"]:
        winner = "무승부"
        print("\n최종 승자는 무승부입니다! 두 사용자 모두 뛰어난 건강 점수를 보여주었습니다.")
    elif len(sorted_users) > 0:
        winner = sorted_users[0][0]
        print(f"\n최종 승자는 **{winner}**님 입니다! 축하드립니다!")
    else:
        winner = "없음 (데이터 부족)"
        print("\n분석할 사용자 데이터가 없습니다.")

    final_output = user_results.copy()
    final_output["승자"] = winner

    # print("\n--- 전체 결과 (JSON 형식) ---")
    print(json.dumps(final_output, indent=2, ensure_ascii=False))
```
