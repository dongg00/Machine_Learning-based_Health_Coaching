```python
import pandas as pd

# 파일 경로
file_csv = "C:/건강검진/health_checkup/JDM/2012년~2023년/건강조사_2012~2022_통합.csv"
file_sas = "C:/건강검진/health_checkup/JDM/2012년~2023년/hn23_all.sas7bdat"

# CSV와 SAS 파일 읽기
df_csv = pd.read_csv(file_csv)
df_sas = pd.read_sas(file_sas, format='sas7bdat', encoding='euc-kr')

# 병합 키 자동 확인
common_cols = list(set(df_csv.columns) & set(df_sas.columns))

# 공통 컬럼으로 병합 시도 (inner join 또는 outer join 선택 가능)
if common_cols:
    df_merged = pd.merge(df_csv, df_sas, on=common_cols, how='outer')
else:
    df_merged = pd.concat([df_csv, df_sas], axis=0, ignore_index=True)

# 결과 미리보기
df_merged.head()

# CSV 저장
df_merged.to_csv('건강조사_2012~2023_통합.csv', index=False, encoding='utf-8-sig')

print('\n✅ CSV 저장 완료: 건강조사_2012~2023_통합.csv')
```

```python
import pandas as pd
import os

# 파일이 있는 폴더 경로 (필요하면 절대경로로 바꿔주세요)
folder_path = "C:/건강검진/health_checkup/JDM/2022년~2023년/"  # 현재 폴더 기준

# 연도 목록: 2012년 ~ 2022년
years = range(12, 23)

# 데이터프레임 모으는 리스트
df_list = []

# 파일 하나씩 처리
for year in years:
    filename = f'hn{year:02d}_all.sas7bdat'  # 예: hn12_all.sas7bdat
    filepath = os.path.join(folder_path, filename)
    
    if os.path.exists(filepath):
        print(f'✅ 불러오는 중: {filename}')
        df = pd.read_sas(filepath, format='sas7bdat')
        df['source_year'] = 2000 + year  # 연도 컬럼 추가
        df_list.append(df)
    else:
        print(f'⚠️ 파일 없음: {filename}')

# 병합
merged_df = pd.concat(df_list, ignore_index=True)

# CSV 저장
merged_df.to_csv('건강조사_2012~2022_통합.csv', index=False, encoding='utf-8-sig')

print('\n✅ CSV 저장 완료: 건강조사_2012~2022_통합.csv')
```
