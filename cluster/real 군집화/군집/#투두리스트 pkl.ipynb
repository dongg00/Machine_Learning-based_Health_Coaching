```python
import numpy as np
import pandas as pd
import json
import random
import pickle
from sklearn.preprocessing import StandardScaler

# Custom JSON encoder for NumPy data
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, (np.integer, np.int32, np.int64)):
            return int(obj)
        elif isinstance(obj, (np.floating, np.float32, np.float64)):
            return float(obj)
        elif isinstance(obj, (np.ndarray,)):
            return obj.tolist()
        return super(NumpyEncoder, self).default(obj)

# Load preprocessed data and scalers
with open('health_service_data.pkl', 'rb') as f:
    saved_data = pickle.load(f)
scalers = saved_data['scalers']
precalculated_means = saved_data['precalculated_means']

# Define precalculated_means (if not loaded from .pkl, include it here as in your original code)
# [Include the precalculated_means dictionary here if you didn't save it in the .pkl file]

# Define health_messages (same as in your original code)
health_messages = {
    'HE_HP2': {
        1: {
            '운동': [
                "매일 6,000보 이상 걷기"
            ],
            '식사': [
                "매일 채소 300g 이상 섭취하기",
                "매일 채소 7접시 이상 (한 끼니 2가지 이상) 섭취하기",
                "매일 단백질 50g 이상 섭취하기",
                "매일 나트륨 2g 이하로 제한하기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "하루 7시간 이상 충분한 수면을 취하기"
            ]
        },
        2: {
            '운동': [
                "유산소 운동과 더불어 등장성 또는 등척성 운동을 병행하기",
                "매일 7,000보 이상 걷기",
                "명상 또는 산책 10분 하기"
            ],
            '식사': [
                "매일 채소 350g 이상 먹기",
                "매일 채소 8접시 이상 (한 끼니 2-3가지 이상) 섭취하기",
                "매일 단백질 55g 이상 섭취하기",
                "매일 나트륨 1.8g 이하 유지하기"
            ],
            '흡연': [],
            '음주': [],
            '기타': []
        },
        3: {
            '운동': [
                "유산소 운동과 더불어 등척성 운동하기"
            ],
            '식사': [
                "채소 400g 이상 먹기",
                "매 끼니마다 채소 9접시 이상 (한 끼니 3가지 이상) 섭취하기",
                "매일 단백질 60g 이상 섭취하기",
                "매일 나트륨 1.5g 이하로 줄이기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "매일 7시간 이상 수면하기"
            ]
        },
        4: {
            '운동': [
                "유산소 운동과 더불어 등장성 또는 등척성 운동을 병행하기",
                "매일 30분 이상 줄넘기 하기",
                "매일 10000보 이상 꼭 걷기",
                "매일 산책 30분 이상 하기"
            ],
            '식사': [
                "매일 채소 500g 이상 먹기",
                "매일 채소 10접시 이상 (한 끼니 4가지 이상) 섭취하기",
                "매일 단백질 65g 이상 섭취하기",
                "매일 나트륨 1.2g 이하 지키기",
                "잡곡밥 먹기",
                "매끼 잡곡밥 2/3~1공기 정도 섭취하기",
                "삼겹살이나 갈비 등의 고지방 육류는 피하기",
                "살코기, 생선, 잡곡을 제외한 닭고기, 콩 두부 등으로 단백질 섭취하기",
                "국물 요리는 건더기 위주로 섭취하기",
                "중간 크기의 사과 1개 먹기",
                "중간 크기의 토마토 2개 먹기",
                "매일 저지방 우유 1컵 마시기",
                "매일 요구르트 1컵 섭취하기",
                "튀김, 전, 볶음요리 먹지 않기",
                "찜, 굽기 등의 조리법 이용하기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "명상과 호흡 각 10분 하기",
                "하루 7시간 이상 충분한 수면 취하기",
                "명상 30분 이상 하기"
            ]
        }
    },
    'HE_DM_HbA1c2': {
        1: {
            '운동': [
                "매일 30분 이상 유산소 운동하기",
                "매일 30분 이상 걷기, 조깅, 수영 등 규칙적으로 하기",
                "하루 30분 이상, 최소 7,000보 걷기"
            ],
            '식사': [
                "매 끼니 잡곡밥, 아침, 단백질 위주의 균형 잡힌 식사 하기",
                "매일 채소 7접시 이상 (한 끼니 2가지 이상) 먹기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "하루 7시간 이상 충분한 수면을 취하기"
            ]
        },
        2: {
            '운동': [
                "하루 30분 이상 빠르게 걷기",
                "하루 30분 이상 자전거 타기",
                "매일 최소 9,000보 걷기",
                "식사 후 10분 정도의 가벼운 산책하기"
            ],
            '식사': [
                "매 끼니 채소, 과일, 통곡물 위주의 식사하기",
                "매일 채소 9접시 이상 (한 끼니 3가지 이상) 먹기",
                "식사를 20분 이상 천천히 하기",
                "한 끼에 적어도 2-3가지의 채소 반찬 먹기",
                "채소가 주재료인 메뉴를 한 가지 이상 고르기",
                "육류 메뉴의 경우 2인분 이상의 과식 피하기",
                "외식 전 하루 식사량의 1/3이 넘지 않도록 하기",
                "외식 시, 30분 이상 천천히 식사하기"
            ],
            '흡연': [],
            '음주': [
                "하루 소주 반 잔 이하",
                "한 번 마실 때 남성은 2잔, 여성은 1잔을 넘지 않도록 하기"
            ],
            '기타': [
                "하루 7시간 이상 충분한 수면"
            ]
        },
        3: {
            '운동': [
                "하루 60분 이상, 최소 10,000보 걸으"
            ],
            '식사': [
                "매일 채소 10접시 이상 (한 끼니 4가지 이상) 먹기",
                "아침 일어나자 2시간 안에 아침식사",
                "식사 간격은 최대 6시간 이내로 하기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "매일 식사 전후로 자가 혈당 측정 및 기록하기",
                "매 운동 전후로 자가 혈당 측정 및 기록하기",
                "매 취침 전 자가 혈당 측정 및 기록하기"
            ]
        }
    },
    'HE_obe2': {
        1: {
            '운동': [
                "식후 30분 정도 가볍게 앉아있기",
                "매일 30분 동안 가벼운 유산소 운동하기",
                "매일 20분 동안 가벼운 근력 운동하기"
            ],
            '식사': [
                "매일 단백질 60g 이상 섭취하기",
                "매일 채소 반찬을 최소 2가지 이상 포함하기",
                "매일 과일 1개 이상 섭취하기",
                "끼니 사이에 건강한 간식을 1회 이상 추가하기",
                "하루 1.5L 이상 물 마시기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "하루 7시간 이상 충분한 수면을 취하기",
                "매일 10분 동안 명상 등 스트레스 해소하기"
            ]
        },
        2: {
            '운동': [
                "매일 7000보 이상 걷기",
                "매일 30분 이상 빠르게 걷기",
                "수영 등 중급 이상의 유산소 운동하기",
                "엘리베이터 대신 계단으로 다니기",
                "대중교통 대신 자전거로 출퇴근하기"
            ],
            '식사': [
                "매 끼니에 탄수화물, 단백질, 지방을 균형 있게 포함하기",
                "매일 채소 반찬을 최소 3가지 이상 포함하기",
                "매일 하루 과일 1개 이상 섭취하기",
                "하루 1.5L 이상 물 마시기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "매일 10분 동안 명상 등 스트레스 해소하기",
                "매일 7시간 이상 충분한 수면을 취하기"
            ]
        },
        3: {
            '운동': [
                "식후 30분 정도 가볍게 앉아있기",
                "매일 30분 동안 가벼운 유산소 운동하기",
                "매일 20분 동안 가벼운 근력 운동하기",
                "엘리베이터 대신 계단으로 다니기",
                "대중교통 대신 자전거로 출퇴근하기"
            ],
            '식사': [
                "매일 단백질 60g 이상 섭취하기",
                "매일 채소 반찬을 최소 2가지 이상 포함하기",
                "끼니 사이에 건강한 간식을 추가하기",
                "하루 1.5L 이상 물 마시기"
            ],
            '흡연': [],
            '음주': [],
            '기타': [
                "하루 7시간 이상 충분한 수면을 취하기",
                "매일 10분 동안 명상 등으로 스트레스 해소하기"
            ]
        },
        4: {
            '운동': [
                "매일 40분 이상 빠르게 걷기",
                "매일 8000보 이상 걷기",
                "엘리베이터 대신 계단으로 다니기",
                "대중교통 대신 자전거로 출퇴근하기"
            ],
            '식사': [
                "하루 세 끼를 정해진 시간에 섭취하기",
                "매일 채소 반찬을 3가지 이상 충분히 섭취하기",
                "매일 과일 1개 이내로 섭취하기",
                "설탕이 많은 음료 대신 물이나 차를 하루 2L 이상 마시기",
                "저녁 식사는 잠들기 최소 3시간 전에 마치기",
                "식사 전에 물 1컵(200ml) 마시기",
                "국물 없는 식사하기"
            ],
            '흡연': [
                "니코틴 껌을 1회 1개 30분간 씹어서 흡연 욕구 해소하기"
            ],
            '음주': [],
            '기타': [
                "매일 7시간 이상 충분한 수면을 취하기",
                "매일 10분 명상으로 스트레스 해소하기"
            ]
        },
        5: {
            '운동': [
                "매일 50분 빠르게 걷기",
                "매일 50분 경사 진 곳에서 걷기",
                "수영 등 중급 이상의 유산소 운동하기",
                "매일 40분 이상 강도 높은 근력 운동하기 (예: 중량 운동)",
                "앉아있는 시간을 줄이고 자주 스트레칭하기",
                "매일 10000보 이상 걷기"
            ],
            '식사': [
                "매일 식이섬유가 풍부한 채소를 500g 이상 섭취하기",
                "매일 해조류와 통곡물을 충분히 섭취하기",
                "매일 과일 1회 1/2개 이내로 먹기",
                "고칼로리 음료, 디저트, 패스트푸드, 튀김류를 멀리하기",
                "저지방 단백질(닭가슴살, 흰살 생선, 콩류)로 먹기",
                "식사 중에는 물을 마시지 않기",
                "식사 30분 전후로 물 2컵(400ml) 마시기",
                "저녁 식사는 잠들기 최소 4시간 전에",
                "모든 야식과 간식은 절대 먹지 않기"
            ],
            '흡연': [],
            '음주': [
                "음주 시 1잔 미만으로 마시기"
            ],
            '기타': []
        },
        6: {
            '운동': [
                "매일 60분 빠르게 걷기",
                "매일 40분 관절에 부담없는 운동하기",
                "매일 40분 근력 운동을 점진적으로 강도를 높이기",
                "하루에 앉아있는 시간을 4시간 미만으로 줄이기",
                "매일 12000보 이상 걷기"
            ],
            '식사': [
                "정해진 식사량을 엄격히 준수하기",
                "탄수화물 섭취량을 하루 100g 이하로 제한하기",
                "단백질(닭가슴살, 생선, 두부 등) 섭취량을 체중 kg당 1.2g 이상으로",
                "매일 식이섬유가 풍부한 채소를 600g 이상",
                "매일 해조류, 통곡물을",
                "과일은 당분이 적은 종류로 하루에 1/2개 이내",
                "모든 가공식품, 당류, 튀김류, 패스트푸드를 먹지 않기",
                "저지방 단백질(닭가슴살, 흰살 생선, 콩류) 위주의 식단을 구성하기",
                "식사 중간에 물을 마시지 않기",
                "식사 30분 전후로 물 2컵(400ml)",
                "저녁 식사는 잠들기 최소 5시간 전에 마치기",
                "모든 야식과 간식은 절대 먹지 않기"
            ],
            '흡연': [],
            '음주': [
                "음주 시 1회 1/2잔 미만"
            ],
            '기타': [
                "매일 8시간 이상의 충분하고 질 높은 수면"
            ]
        }
    }
}

# Clustering-related functions
def determine_cluster_with_precalculated_means(user_data_weekly, condition_value, scaler, precalculated_means_dict, selected_cols, condition_type):
    condition_value_int = int(condition_value)

    if condition_type not in precalculated_means_dict or condition_value_int not in precalculated_means_dict[condition_type]:
        return None, None

    cluster_means_precalculated_text_names = precalculated_means_dict[condition_type][condition_value_int].copy()

    text_to_internal_col_mapping = {
        '1주일 간 음주 빈도': 'BD1_11',
        '하루 평균 흡연량': 'tobacco',
        '1주일 간 걷기 일수': 'BE3_31',
        '1주일 간 아침식사 빈도': 'L_BR_FQ'
    }

    cluster_means_precalculated_internal_names = pd.DataFrame(index=cluster_means_precalculated_text_names.index)
    for internal_col in selected_cols:
        text_col = next((k for k, v in text_to_internal_col_mapping.items() if v == internal_col), None)
        if text_col and text_col in cluster_means_precalculated_text_names.columns:
            cluster_means_precalculated_internal_names[internal_col] = cluster_means_precalculated_text_names[text_col]
        else:
            cluster_means_precalculated_internal_names[internal_col] = np.nan

    if cluster_means_precalculated_internal_names.isnull().values.any():
        return None, None

    scaled_precalculated_centroids = scaler.transform(cluster_means_precalculated_internal_names[selected_cols])

    user_data_df = pd.DataFrame(
        [user_data_weekly],
        columns=['BD1_11', 'tobacco', 'BE3_31', 'L_BR_FQ']
    )
    user_data_scaled_for_predict = scaler.transform(user_data_df)

    closest_cluster = np.argmin(np.linalg.norm(scaled_precalculated_centroids - user_data_scaled_for_predict, axis=1)) + 1
    cluster_means = precalculated_means_dict[condition_type][condition_value_int].iloc[closest_cluster - 1]
    return closest_cluster, cluster_means

def calculate_health_score(cluster_mean, condition_type):
    weights = {
        'HE_HP2': {
            'tobacco': 0.383886,
            'BE3_31': 0.277795,
            'BD1_11': 0.188051,
            'L_BR_FQ': 0.150268
        },
        'HE_DM_HbA1c2': {
            'tobacco': 0.362728,
            'BE3_31': 0.275424,
            'BD1_11': 0.199061,
            'L_BR_FQ': 0.162787
        },
        'HE_obe2': {
            'tobacco': 0.398547,
            'BE3_31': 0.265374,
            'BD1_11': 0.198054,
            'L_BR_FQ': 0.138025
        }
    }

    if condition_type not in weights:
        raise ValueError(f"알 수 없는 condition_type: {condition_type}")

    w = weights[condition_type]
    
    score = (
        -cluster_mean['1주일 간 음주 빈도'] * w['BD1_11']
        -cluster_mean['하루 평균 흡연량'] * w['tobacco']
        +cluster_mean['1주일 간 걷기 일수'] * w['BE3_31']
        +cluster_mean['1주일 간 아침식사 빈도'] * w['L_BR_FQ']
    )
    return score

def calculate_cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0
    return dot_product / (norm_vec1 * norm_vec2)

def find_healthier_and_similar_cluster(user_data_raw, user_cluster_means, condition_value, precalculated_means_dict, condition_type):
    condition_value_int = int(condition_value)

    if condition_type not in precalculated_means_dict or condition_value_int not in precalculated_means_dict[condition_type]:
        return None, None

    cluster_means_all = precalculated_means_dict[condition_type][condition_value_int]
    health_scores = cluster_means_all.apply(lambda row: calculate_health_score(row, condition_type), axis=1)

    user_cluster_score = calculate_health_score(user_cluster_means, condition_type)

    user_data_vector_raw = np.array([
        user_data_raw[0],  # 음주 빈도
        user_data_raw[1],  # 흡연량
        user_data_raw[2],  # 걷기 일수
        user_data_raw[3]   # 아침식사 빈도
    ])

    healthier_cluster_index = None
    max_cosine_similarity = -1
    for idx in cluster_means_all.index:
        current_cluster_mean = cluster_means_all.loc[idx]
        current_score = calculate_health_score(current_cluster_mean, condition_type)
        if current_score > user_cluster_score:
            cluster_mean_vector_raw = current_cluster_mean[['1주일 간 음주 빈도', '하루 평균 흡연량', '1주일 간 걷기 일수', '1주일 간 아침식사 빈도']].values
            sim = calculate_cosine_similarity(user_data_vector_raw, cluster_mean_vector_raw)
            if sim > max_cosine_similarity:
                max_cosine_similarity = sim
                healthier_cluster_index = idx

    if healthier_cluster_index is not None:
        return healthier_cluster_index, cluster_means_all.loc[healthier_cluster_index]
    return None, user_cluster_means

def determine_he_hp2(sbp, dbp):
    if pd.isna(sbp) or pd.isna(dbp):
        return np.nan
    if sbp >= 140 or dbp >= 90:
        return 4
    elif 130 <= sbp <= 139 or 80 <= dbp <= 89:
        return 3
    elif 120 <= sbp <= 129 and dbp < 80:
        return 2
    else:
        return 1

def determine_he_dm_hba1c2(glu):
    if pd.isna(glu):
        return np.nan
    if glu <= 99:
        return 1
    elif 100 <= glu <= 125:
        return 2
    else:
        return 3

def determine_he_obe2(bmi):
    if pd.isna(bmi):
        return np.nan
    if bmi <= 18.5:
        return 1
    elif 18.5 < bmi <= 22.9:
        return 2
    elif 22.9 < bmi <= 24.9:
        return 3
    elif 24.9 < bmi <= 29.9:
        return 4
    elif 29.9 < bmi <= 34.9:
        return 5
    else:
        return 6

def integrated_health_service(user_data):
    try:
        name = user_data.get('user_name', 'Unknown')
        required_keys = ['HE_sbp1', 'HE_dbp1', 'HE_glu', 'HE_BMI', 'weekly_alcohol', 'daily_smoking', 'weekly_exercise', 'daily_breakfast']
        for key in required_keys:
            if key not in user_data or user_data[key] is None:
                return {"error": f"{key} 값이 누락되었거나 None입니다."}

        try:
            sbp = float(user_data['HE_sbp1'])
            dbp = float(user_data['HE_dbp1'])
            glu = float(user_data['HE_glu'])
            bmi = float(user_data['HE_BMI'])
            weekly_alcohol = float(user_data['weekly_alcohol'])
            daily_smoking = float(user_data['daily_smoking'])
            weekly_exercise = float(user_data['weekly_exercise'])
            daily_breakfast = float(user_data['daily_breakfast'])
        except (ValueError, TypeError) as e:
            return {"error": f"입력값 변환 실패: {str(e)}"}

        disease_labels = {
            'HE_HP2': {1: "정상", 2: "주의 혈압", 3: "고혈압 전단계", 4: "고혈압"},
            'HE_DM_HbA1c2': {1: "정상", 2: "당뇨 전단계", 3: "당뇨"},
            'HE_obe2': {1: "저체중", 2: "정상", 3: "과체중", 4: "비만", 5: "고도비만", 6: "초고도비만"},
        }

        disease_funcs = {
            'HE_HP2': lambda sbp, dbp: determine_he_hp2(sbp, dbp),
            'HE_DM_HbA1c2': lambda glu, _: determine_he_dm_hba1c2(glu),
            'HE_obe2': lambda bmi, _: determine_he_obe2(bmi),
        }

        user_inputs = [weekly_alcohol, daily_smoking, weekly_exercise, daily_breakfast]
        categories = ['1주일 간 음주 빈도', '하루 평균 흡연량', '1주일 간 걷기 일수', '1주일 간 아침식사 빈도']
        selected_cols = ['BD1_11', 'tobacco', 'BE3_31', 'L_BR_FQ']

        # Actionable message lists
        alcohol_reduce_messages = [
            "하루 음주량을 {abs_diff_alcohol}잔 줄이기",
            "하루 음주량을 {abs_diff_alcohol}잔 줄여 더 건강해지기",
            "현재 하루 음주량에서 {abs_diff_alcohol}잔 정도 줄이기",
            "하루 음주량을 {abs_diff_alcohol}잔 감소시키기"
        ]

        smoking_reduce_messages = [
            "하루 흡연량을 {abs_diff_smoking}개비 줄여 금연에 도전하기",
            "하루 흡연량을 {abs_diff_smoking}개비 줄여 건강한 삶을 살기",
            "하루 현재 흡연량에서 {abs_diff_smoking}개비 정도 줄이기",
            "하루 흡연량을 {abs_diff_smoking}개비 줄여 건강을 개선하기",
            "하루 흡연량을 {abs_diff_smoking}개비 감소시키기"
        ]

        exercise_increase_messages = [
            "하루 {abs_diff_exercise_minutes}분 더 걸어 활기찬 하루 보내기",
            "하루 {abs_diff_exercise_minutes}분 더 걸어 건강 증진하기",
            "목표 걷기 시간에 도달하려면 하루 {abs_diff_exercise_minutes}분 더 걷기",
            "지금보다 하루 {abs_diff_exercise_minutes}분 더 걸어 건강해지기",
            "하루 {abs_diff_exercise_minutes}분 더 걸어 건강한 습관 만들기"
        ]

        breakfast_increase_messages = [
            "하루 아침식사 빈도를 {abs_diff_breakfast}회 더 늘려 건강한 하루 시작하기",
            "규칙적인 아침 식사를 위해 하루 {abs_diff_breakfast}회 더 챙겨 먹기",
            "하루 아침 식사를 {abs_diff_breakfast}회 늘려 하루의 활력 얻기",
            "하루 아침 식사 빈도를 {abs_diff_breakfast}회 늘려 건강한 식습관 만들기",
            "하루 아침식사 횟수를 {abs_diff_breakfast}회 늘려 균형 잡힌 영양 섭취하기"
        ]

        messages_pool = {'운동': [], '식사': [], '흡연': [], '음주': [], '기타': []}
        category_mapping = {
            'alcohol': '음주',
            'smoking': '흡연',
            'exercise': '운동',
            'breakfast': '식사'
        }

        # Process each disease condition
        for disease_code in ['HE_HP2', 'HE_DM_HbA1c2', 'HE_obe2']:
            if disease_code == 'HE_HP2':
                val = disease_funcs[disease_code](sbp, dbp)
            else:
                val = disease_funcs[disease_code](glu if disease_code == 'HE_DM_HbA1c2' else bmi, None)

            if pd.isna(val):
                continue
            val_int = int(val)

            # Load scaler for the specific condition and value
            scaler = scalers.get(disease_code, {}).get(val_int)
            if scaler is None:
                # Fallback: Use precalculated means if no scaler is available
                cluster_means_df = precalculated_means[disease_code][val_int][categories]
                scaler = StandardScaler()
                scaler.fit(cluster_means_df)

            cluster, cluster_means = determine_cluster_with_precalculated_means(user_inputs, val_int, scaler, precalculated_means, selected_cols, disease_code)
            if cluster is None:
                continue

            target_idx, target_means = find_healthier_and_similar_cluster(user_inputs, cluster_means, val_int, precalculated_means, disease_code)
            if target_idx is None:
                target_idx = cluster
                target_means = cluster_means

            # Calculate differences and prepare actionable messages
            user_values = user_inputs

            # Alcohol frequency
            if user_values[0] > 0:
                diff_alcohol_weekly = user_values[0] - target_means['1주일 간 음주 빈도']
                if diff_alcohol_weekly > 0:
                    abs_diff_alcohol_daily = round(abs(diff_alcohol_weekly) * 6.1 / 7, 2)
                    messages_pool['음주'].append(random.choice(alcohol_reduce_messages).format(abs_diff_alcohol=abs_diff_alcohol_daily))

            # Smoking
            if user_values[1] > 0:
                diff_smoking = user_values[1] - target_means['하루 평균 흡연량']
                if diff_smoking > 0:
                    abs_diff_smoking = round(abs(diff_smoking), 1)
                    messages_pool['흡연'].append(random.choice(smoking_reduce_messages).format(abs_diff_smoking=abs_diff_smoking))

            # Exercise (weekly walking days)
            diff_exercise_weekly = target_means['1주일 간 걷기 일수'] - user_values[2]
            if diff_exercise_weekly > 0:
                abs_diff_exercise_minutes = round((diff_exercise_weekly * 30) / 7, 0)
                if abs_diff_exercise_minutes > 0:
                    messages_pool['운동'].append(random.choice(exercise_increase_messages).format(abs_diff_exercise_minutes=abs_diff_exercise_minutes))

            # Breakfast frequency
            diff_breakfast_weekly = target_means['1주일 간 아침식사 빈도'] - user_values[3]
            if diff_breakfast_weekly > 0:
                abs_diff_breakfast_daily = round(diff_breakfast_weekly / 7, 2)
                messages_pool['식사'].append(random.choice(breakfast_increase_messages).format(abs_diff_breakfast=abs_diff_breakfast_daily))

            # Add default messages
            if disease_code in health_messages and val_int in health_messages[disease_code]:
                for category in messages_pool:
                    if health_messages[disease_code][val_int][category]:
                        messages_pool[category].extend(health_messages[disease_code][val_int][category])

        # Generate final todo list (4 messages, preferably from different categories)
        final_todo_list = []
        used_categories = set()
        existing_messages = set()

        available_categories = ['운동', '식사', '흡연', '음주', '기타']
        random.shuffle(available_categories)
        for category in available_categories:
            if len(final_todo_list) >= 4:
                break
            valid_messages = [msg for msg in messages_pool[category] if msg not in existing_messages]
            if valid_messages:
                selected_message = random.choice(valid_messages)
                final_todo_list.append(selected_message)
                used_categories.add(category)
                existing_messages.add(selected_message)

        while len(final_todo_list) < 4:
            for category in available_categories:
                valid_messages = [msg for msg in messages_pool[category] if msg not in existing_messages]
                if valid_messages:
                    selected_message = codzien.choice(valid_messages)
                    final_todo_list.append(selected_message)
                    existing_messages.add(selected_message)
                    break
                else:
                    break

        final_todo_list = final_todo_list[:4]

        return {
            "user_name": name,
            "todo_list": final_todo_list
        }

    except Exception as e:
        return {"error": str(e)}

if __name__ == "__main__":
    sample_user_data = {
        'user_name': '김미정',
        'HE_sbp1': 135,
        'HE_dbp1': 90,
        'HE_glu': 88,
        'HE_BMI': 28,
        'weekly_alcohol': 4,
        'daily_smoking': 5,
        'weekly_exercise': 4,
        'daily_breakfast': 7
    }

    result = integrated_health_service(sample_user_data)
    print(json.dumps(result, indent=2, ensure_ascii=False, cls=NumpyEncoder))
```
